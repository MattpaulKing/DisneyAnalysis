```python
import numpy as np
import pandas as pd
```

```python
total_gross = pd.read_csv("disney_movies_total_gross.csv")
disney_revenue = pd.read_csv("disney_revenue_1991-2016.csv")
disney_characters = pd.read_csv("disney-characters.csv")
disney_directors = pd.read_csv("disney-director.csv")
disney_actors = pd.read_csv("disney-voice-actors.csv")
```

```python
def tweak_df(dis_df):
    return (dis_df
    .assign(movie_title=dis_df.movie_title.astype("string"), 
            release_date=pd.to_datetime(dis_df.release_date), 
            genre=dis_df.genre.fillna("No Genre").astype('string'), 
            MPAA_rating=dis_df.MPAA_rating.fillna("Not Rated").astype('string'), 
            inflation_adjusted_gross=dis_df.inflation_adjusted_gross.str.replace(",", "", regex=True).str.replace("$","", regex=True).astype('float32'), 
            #director=dis_df.director.astype('string'), 
            year=pd.to_datetime(dis_df.release_date).dt.year
    )
    .drop(columns=['total_gross']))
```

```python
disney_directors = disney_directors.rename(columns={'name': 'movie_title'})
total_gross_director = total_gross.merge(disney_directors, how='left', on='movie_title')
cleaned_df = tweak_df(total_gross)
cleaned_df
```

```python
genre_df = cleaned_df.groupby(['genre', 'MPAA_rating', 'year'])['inflation_adjusted_gross'].sum().reset_index()
genre_df
```
